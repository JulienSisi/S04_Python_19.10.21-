## 2.2.8. Quel département français possède le plus de musées sur son territoire ?

```python
museums['postal_code'].value_counts()
```

```python
museums["departement"] = museums['postal_code'].astype(str).str[0:2]
museums["departement"].value_counts()
```

```python
departement = museums['postal_code'].dropna()
departement = departement.astype(int).astype(str).str.slice(stop=2)
departement
```

```python
departement = museums['postal_code'].dropna()
departement = departement.astype(int).astype(str).str.slice(stop=2)
departement.value_counts().idxmax()
```

### Réponse : Selon six méthodes, l'île de France

## 2.2.9. Quel département français possède le moins de musées sur son territoire ?

```python
departement = museums['postal_code'].dropna()
departement = departement.astype(int).astype(str).str.slice(stop=2)
departement.value_counts().idxmin()
```

### Réponse : Les départements d'outre mer, 987 : Polynésie française. 988 : Nouvelle-Calédonie. 989 : Île de Clipperton.
